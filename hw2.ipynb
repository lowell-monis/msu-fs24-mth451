### Section 2.1 Question 6 (c) - The Bisection Method



```python
import numpy as np

def bisection_method(f, a, b, tol=1e-7):
    max_iter =  round(np.log((b - a) / 1e-5) / np.log(2))
    if f(a) * f(b) >= 0:
        print("The function must have opposite signs at a and b.")
        return None

    for i in range(max_iter):
        # Calculate the midpoint
        c = (a + b) / 2.0

        # If the value at the midpoint is close enough to zero, return c
        if abs(f(c)) < tol:
            return c
        
        # Determine the new interval to use
        if f(a) * f(c) < 0:
            b = c  # Root lies in the left half
        else:
            a = c  # Root lies in the right half

    # If we reach max_iter iterations without finding the root
    print("Maximum iterations reached.")
    return (a + b) / 2.0  # Return the best estimate

# Define the function for which you want to find the root
def f(x):
    return x**2 - 4*x +4 - np.log(x)
    
# Call the bisection method with an initial interval [a, b]
root_1 = round(bisection_method(f, 2, 4), 5)
root_2 = round(bisection_method(f, 1, 2), 5)
print(f"Approximate roots: {root_1, root_2}")
```

### Section 2.3 Question 2 - Newton's Method and its Extensions



```python
import math

def f(x):
    return -x**3 - math.cos(x)

def f_prime(x):
    return -3*x**2 + math.sin(x)

def newtons_method(p0, tol=1e-7, max_iter=100):
    for i in range(max_iter):
        f_p0 = f(p0)
        f_prime_p0 = f_prime(p0)
        
        # Check if the derivative is zero to avoid division by zero
        if f_prime_p0 == 0:
            print(f"Derivative zero at x = {p0}, cannot proceed with Newton's method.")
            return None
        
        # Update p1 using Newton's method formula
        p1 = p0 - f_p0 / f_prime_p0
        
        # If the absolute difference is below the tolerance, return the result
        if abs(p1 - p0) < tol:
            return p1
        
        # Update p0 for the next iteration
        p0 = p1
    
    print("Newton's method did not converge within the maximum number of iterations.")
    return p0

# Example usage
p0 = -1  # Initial guess
root = newtons_method(p0)
print(f"The root found is approximately: {root}")
```
