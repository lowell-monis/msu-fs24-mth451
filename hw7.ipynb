I have done most of the setup of this question manually. The summation, however, may take time, which is why I will be doing it computationally.

I will thus be defining a function for the summation, and for the function.

```python
import numpy as np

def euler_method(f, t0, y0, h, steps):
    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0
    for _ in range(steps):
        y += h * f(t, y)
        t += h
        t_values.append(t)
        y_values.append(y)
    return t_values, y_values

# Problem 1
f1 = lambda t, y: t * np.exp(3 * t) - 2 * y
t0_1, y0_1, h1, steps1 = 0, 0, 0.5, 2
t_values1, y_values1 = euler_method(f1, t0_1, y0_1, h1, steps1)

# Problem 2
f2 = lambda t, y: (np.sin(2 * t) - 2 * t * y) / t**2
t0_2, y0_2, h2, steps2 = 1, 2, 0.25, 4
t_values2, y_values2 = euler_method(f2, t0_2, y0_2, h2, steps2)

print("Problem 1 Results:")
for t, y in zip(t_values1, y_values1):
    print(f"t = {t:.2f}, y = {y:.5f}")

print("\nProblem 2 Results:")
for t, y in zip(t_values2, y_values2):
    print(f"t = {t:.2f}, y = {y:.5f}")
```
