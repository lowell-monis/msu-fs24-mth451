```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = ln(x)
def f(x):
    return np.log(x)

# Cubic Spline Interpolation
def cubic_spline(x, y, num_points=100):
    n = len(x)
    h = np.diff(x)
    a = y[:-1]
    
    # Create matrices for the system of equations
    A = np.zeros((n, n))
    b = np.zeros(n)
    
    # Set up the equations
    A[0, 0] = 1  # Natural spline boundary condition
    for i in range(1, n - 1):
        A[i, i - 1] = h[i - 1]
        A[i, i] = 2 * (h[i - 1] + h[i])
        A[i, i + 1] = h[i]
        b[i] = 3 * ((y[i + 1] - y[i]) / h[i] - (y[i] - y[i - 1]) / h[i - 1])
    
    A[n - 1, n - 1] = 1  # Natural spline boundary condition

    # Solve the system of equations for c
    c = np.linalg.solve(A, b)
    
    # Calculate b and d
    b_coeff = np.zeros(n - 1)
    d_coeff = np.zeros(n - 1)
    
    for i in range(n - 1):
        b_coeff[i] = (y[i + 1] - y[i]) / h[i] - h[i] * (2 * c[i] + c[i + 1]) / 3
        d_coeff[i] = (c[i + 1] - c[i]) / (3 * h[i])
    
    # Generate spline points
    x_spline = np.linspace(x[0], x[-1], num_points)
    y_spline = np.zeros_like(x_spline)
    
    for j in range(num_points):
        # Find the interval for x_spline[j]
        for i in range(n - 1):
            if x[i] <= x_spline[j] <= x[i + 1]:
                # Evaluate the spline at x_spline[j]
                y_spline[j] = (a[i] + 
                               b_coeff[i] * (x_spline[j] - x[i]) + 
                               c[i] * (x_spline[j] - x[i])**2 + 
                               d_coeff[i] * (x_spline[j] - x[i])**3)
                break

    return x_spline, y_spline

# Define the interval and number of points (n=4)
x_points = np.linspace(1, 5, 4)  # n=4 points in the interval [1,5]
y_points = f(x_points)

# Perform cubic spline interpolation
x_spline, y_spline = cubic_spline(x_points, y_points)

# Plot the original data points and the cubic spline interpolation
plt.plot(x_points, y_points, 'o', label='Data points (n=4)')
plt.plot(x_spline, y_spline, label='Cubic Spline Interpolation')
plt.plot(x_spline, f(x_spline), '--', label='Original function f(x) = ln(x)', color='gray')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Cubic Spline Interpolation for f(x) = ln(x)')
plt.show()
```

```python
import numpy as np

def natural_cubic_spline(x, a):
    """
    Computes the natural cubic spline interpolation for the function values given at points x.
    
    Parameters:
        x: array-like, the known data points x0 < x1 < ... < xn
        a: array-like, the function values a0 = f(x0), a1 = f(x1), ..., an = f(xn)
    
    Returns:
        aj, bj, cj, dj: arrays of spline coefficients
    """
    n = len(x) - 1  # Number of intervals
    h = np.diff(x)   # h[i] = x[i+1] - x[i]
    
    # Step 2: Compute alpha
    alpha = np.zeros(n)
    for i in range(1, n):
        alpha[i] = (3 / h[i]) * (a[i+1] - a[i]) - (3 / h[i-1]) * (a[i] - a[i-1])
    
    # Step 3: Set up the system
    l = np.ones(n+1)
    mu = np.zeros(n)
    z = np.zeros(n+1)
    
    # Step 4: Forward elimination
    for i in range(1, n):
        l[i] = 2 * (x[i+1] - x[i-1]) - h[i-1] * mu[i-1]
        mu[i] = h[i] / l[i]
        z[i] = (alpha[i] - h[i-1] * z[i-1]) / l[i]
    
    # Step 5: Back substitution
    c = np.zeros(n+1)
    b = np.zeros(n)
    d = np.zeros(n)
    
    for j in range(n-1, -1, -1):
        c[j] = z[j] - mu[j] * c[j+1]
        b[j] = (a[j+1] - a[j]) / h[j] - h[j] * (c[j+1] + 2 * c[j]) / 3
        d[j] = (c[j+1] - c[j]) / (3 * h[j])
    
    # Output: aj, bj, cj, dj for j = 0, ..., n-1
    return a[:-1], b, c[:-1], d

# Example usage:
x = np.linspace(1, 5, 5)  # Known data points
a = f(x)  # Function values at those points
aj, bj, cj, dj = natural_cubic_spline(x, a)

# Print the spline coefficients
for j in range(len(aj)):
    print(f"Spline {j}: aj = {aj[j]}, bj = {bj[j]}, cj = {cj[j]}, dj = {dj[j]}")
```

```python
import numpy as np

def clamped_cubic_spline(x, y, FPO, FPN):
    """
    Computes the clamped cubic spline interpolation for the function values given at points x,
    with clamped boundary conditions for the derivatives at the endpoints.

    Parameters:
        x: array-like, the known data points x0 < x1 < ... < xn
        a: array-like, the function values a0 = f(x0), a1 = f(x1), ..., an = f(xn)
        FPO: float, the derivative of the function at x0 (S'(x0) = f'(x0))
        FPN: float, the derivative of the function at xn (S'(xn) = f'(xn))

    Returns:
        aj, bj, cj, dj: arrays of spline coefficients
    """
    n = len(x) - 1
    h = np.diff(x)  # intervals between points
    a = y.copy()  # spline coefficients (function values)

    # Step 2: Calculate alpha
    alpha = np.zeros(n+1)
    alpha[0] = 3 * (a[1] - a[0]) / h[0] - 3 * FPO  # clamped condition at x0
    alpha[n] = 3 * FPN - 3 * (a[n] - a[n-1]) / h[n-1]  # clamped condition at xn

    for i in range(1, n):
        alpha[i] = 3 * (a[i+1] - a[i]) / h[i] - 3 * (a[i] - a[i-1]) / h[i-1]

    # Step 3: Solve the tridiagonal system
    l = np.ones(n+1)
    mu = np.zeros(n+1)
    z = np.zeros(n+1)

    l[0] = 2 * h[0]
    mu[0] = 0.5
    z[0] = alpha[0] / l[0]

    for i in range(1, n):
        l[i] = 2 * (x[i+1] - x[i-1]) - h[i-1] * mu[i-1]
        mu[i] = h[i] / l[i]
        z[i] = (alpha[i] - h[i-1] * z[i-1]) / l[i]

    l[n] = h[n-1] * (2 - mu[n-1])
    z[n] = (alpha[n] - h[n-1] * z[n-1]) / l[n]

    # Step 4: Back substitution to solve for c, b, d
    c = np.zeros(n+1)
    b = np.zeros(n)
    d = np.zeros(n)

    c[n] = z[n]

    for j in range(n-1, -1, -1):
        c[j] = z[j] - mu[j] * c[j+1]
        b[j] = (a[j+1] - a[j]) / h[j] - h[j] * (c[j+1] + 2 * c[j]) / 3
        d[j] = (c[j+1] - c[j]) / (3 * h[j])

    # Return the coefficients a, b, c, d
    return a, b, c, d


# Example usage:
x = np.linspace(1, 5, 5)  # Known data points
a = f(x)  # Function values at those points
FPO = f(x[0])
FPN = f(x[-1])

aj, bj, cj, dj = clamped_cubic_spline(x, a, FPO, FPN)

# Print the spline coefficients
for j in range(len(bj)):
    print(f"Spline {j}: aj = {aj[j]}, bj = {bj[j]}, cj = {cj[j]}, dj = {dj[j]}")
```
