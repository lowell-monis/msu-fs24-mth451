### Section 4.4 Question 2 (c) - Composite Numerical Integration




I have done most of the setup of this question manually. The summation, however, may take time, which is why I will be doing it computationally.

I will thus be defining a function for the summation, and for the function.

```python
import numpy as np

def summation(function, n, a, b):
    intervals = np.linspace(a, b, n+1) # from x0 to x8
    total = 0
    for i in range(1, n):
        total += function(intervals[i])
    return total

def f(x):
    return (np.sin(x))**2 -2*x*np.sin(x) + 1
```

I want to find

$$\sum^{n-1}_{j=1} f(x_j) = \sum^{n-1}_{j=1} (\sin^2 x -2x\sin x +1)$$

to complete the integration approximation.

```python
summation(f, 8, 0.75, 1.75)
```
