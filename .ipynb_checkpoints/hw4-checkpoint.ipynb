### Section 3.5 Question 2 - Cubic Spline Interpolation



Equations have been set up. We will be using the `sympy` module to solve these equations and find the coefficients.

```python
from sympy import *

# Define the variables
a0, b0, c0, d0, a1, b1, c1, d1 = symbols('a0 b0 c0 d0 a1 b1 c1 d1')

eq1 = Eq(a0, 0.0)
eq2 = Eq(b0, 1.0)
eq3 = Eq(a1, 1.0)
eq4 = Eq(b1+ (2*c1) + 3*d1, 1.0)
eq5 = Eq(c0 + d0, 0.0)
eq6 = Eq(c0 + (3*d0), c1)
eq7 = Eq((2*c0) + 3*(d0) + 1, b1)
eq8 = Eq(b1 + c1 + d1, 1)

# Solving the above system
solutions = solve([eq1, eq2, eq3, eq4, eq5, eq6, eq7, eq8], [a0, b0, c0, d0, a1, b1, c1, d1])
solutions
```

### Section 3.5 Question 3 (c) - Cubic Spline Interpolation





We use the same method.

```python
# Define the variables
a0, b0, c0, d0, a1, b1, c1, d1 = symbols('a0 b0 c0 d0 a1 b1 c1 d1')

# Given conditions
eq1 = Eq(a0, -0.02475)
eq2 = Eq(a0 + b0*(-0.25 + 0.5) + c0*(-0.25 + 0.5)**2 + d0*(-0.25 + 0.5)**3, 0.3349375)
eq3 = Eq(a1, 0.3349375)
eq4 = Eq(a1 + b1*(0 + 0.25) + c1*(0 + 0.25)**2 + d1*(0 + 0.25)**3, 1.101)

# Natural spline boundary conditions
eq5 = Eq(2*c0, 0)
eq6 = Eq(2*c1 + 6*d1*(0 + 0.25), 0) 

# Continuity conditions at x = -0.25
eq7 = Eq(b0 + 2*c0*(-0.25 + 0.5) + 3*d0*(-0.25 + 0.5)**2, b1) 
eq8 = Eq(2*c0 + 6*d0*(-0.25 + 0.5), 2*c1) 

# Solving the above system
solutions = solve([eq1, eq2, eq3, eq4, eq5, eq6, eq7, eq8], [a0, b0, c0, d0, a1, b1, c1, d1])
solutions
```

### Section 3.6 Question 1 (a) - Parametric Curves





I have used `matplotlib` here along with mathematical defined functions to plot the curves.

```python
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    return (-10)*(t**3) + 14*(t**2) + t

def y(t):
    return (-2)*(t**3) + 3*(t**2) + t

t_values = np.linspace(-1,1)
x_values = x(t_values)
y_values = y(t_values)

plt.plot(x_values, y_values)
```
